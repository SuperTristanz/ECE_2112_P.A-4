### Problem 1: Data Wrangling

```python
# importing panda Library
import pandas as pd
```

```python
# reading the Excel 
df = pd.read_excel('board2.xlsx')
df
```

#### a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# locating the name Name, GEAS, and electronics 70 up 
instru = df.loc[(df['Hometown'] == 'Luzon') & (df['Electronics'] > 70), ['Name', 'GEAS', 'Electronics']]
instru
```

```python
df['Average'] = df[['Math','Electronics','GEAS','Communication']].mean(axis=1)
```

```python
mindy = df.loc[(df['Hometown']== 'Mindanao') & (df['Gender'] == 'Female') &(df['Average']>55),['Name','Track','Electronics','Average']]
mindy
```

### Problem 2: Data Visualization

```python
#import Data visualization library
import matplotlib.pyplot as plt
```

```python
#determining the size of the table
plt.figure(figsize=(5, 5))
# making the Bar graph
track = plt.bar(df['Track'], df['Average'])
# lebeling the x, y and the title of the graph
plt.xlabel('Track')
plt.ylabel('Average')
plt.title('Average by Track')
# showing of the graph
plt.show()
```

### It is said on the Table that people who took Microelectronics tracks perform better on boards, followed by Communication Track, then Instrumentation Track

## Classification By Gender

```python
#setting the figure size
plt.figure(figsize = (5,5))
track = plt.bar(df['Gender'], df['Average'])

#Labeling the genders and average and the title
plt.xlabel('Gender')
plt.ylabel('Average')
plt.title('Average by Gender')

#showing the table
plt.show()
```

### The table displays that Female students perform better on board exams that Male students

## Hometown

```python
# figuring the size of the table
plt.figure(figsize = (5,5))
#making the bar
track = plt.bar(df['Hometown'], df['Average'])
#Labeling the  x , y and putting a title
plt.xlabel('Hometown')
plt.ylabel('Average')
plt.title('Average Score by Hometown')
#Showing the table
plt.show()
```

### The Table displays that People who are hometown is on luzon perferm better followed by Visayas and then Mindanao on Board exam
